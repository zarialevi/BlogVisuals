```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("dark")
sns.set_palette(palette="husl")
tips = sns.load_dataset("diamonds")
plt.ylim(3300, 4675)
ax = sns.barplot(x="cut", y="price", data=tips)

plt.savefig('img1.png')
```
